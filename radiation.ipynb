```python
import numpy as np
import matplotlib.pyplot as plt
data = np.genfromtxt(fname='radiation.txt')
z = data[:,1]
p = data[:,2]
t = data[:,3]
E_down = data[:,4]
E_up = data[:,5]
plt.plot(E_down,p,'b',label='E_down ($w/m^2$)')
plt.plot(E_up,p,'r',label='E_up ($w/m^2$)')
plt.plot(t,p,'y',label='temp (k)')
plt.ylim(1000,0)
plt.legend()
plt.ylabel('Pressure')plt.plot()
```
